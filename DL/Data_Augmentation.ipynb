```python
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('/content/train/IMG-20240815-WA0007.jpg', target_size=(200, 200))
```

```python
img
```

```python
type(img)
```

### Transformations

```python
datagen = ImageDataGenerator(
    rotation_range = 30,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    fill_mode = 'constant' # 'nearest', 'reflect', 'cosntant'
)
```

### Type conversion from PIL to Array

```python
img = image.img_to_array(img)
```

```python
img.shape
```

```python
type(img)
```

### Adding batch element

```python
input_batch = img.reshape(1, 200, 200, 3)
```

### Generation

```python
i = 0
# for output in dataget.flow_from_directory() -> entire folder
# for single image datagen.flow()
for output in datagen.flow(input_batch, batch_size = 1, save_to_dir='aug'):
  i = i+1
  if i == 10:
    break
```
