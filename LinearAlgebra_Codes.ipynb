### Dot or Inner product or Scalar product

```python
p = np.array([-4, 2, -2])
q = np.array([-1, 3, 4])
print("Dot product: ", np.dot(p, q))
print("Inner product p and q: ", np.inner(p, q))
# p.q = q.p
print("Inner product q and p: ", np.inner(q, p))

print("@ product: ", p@q)

ones = np.array([1, 1, 1])
print("Dot product of p with 1: ", np.dot(ones, p))

# inner product divided by scalar
print("inner product divided by scalar", np.inner(ones, q) / q.shape[0])
```

### Euclidean Norm and Euclidean distance of an n-vector

```python
e = np.array([5, 4, 3, 1, -1, -3, -5])
print("Euclidean norm: ", np.linalg.norm(e))

# Root mean square
print("Root mean square: ", np.linalg.norm(e) / np.sqrt(e.shape[0]))
print("Root mean square: ", np.linalg.norm(e) / np.sqrt(len(e)))

v = np.array([-3, -2, -1, 0, 1, 2, 3])
print("Euclidean distance v - e: ", np.linalg.norm(v-e))
print("Euclidean distance e - v: ", np.linalg.norm(e-v))
```

## üîπ Vector Independence & Dependence

### (Explained using Angle, Plane, Determinant & Inverse)

---

## 1. **Vector Independence**

A set of vectors is **linearly independent** if **no vector can be written as a combination of the others**.

### üëâ Simple meaning

Each vector provides a **new direction**.

### üëâ Geometric meaning (plane concept)

* In **2D**: two independent vectors span the **entire plane**.
* In **3D**: three independent vectors span the **entire space**.

They do **not** lie on the same line or the same plane (for 3D).

### üëâ Angle interpretation

* If the angle between two vectors is **not 0¬∞ or 180¬∞**, they are independent.
* **Orthogonal vectors (90¬∞)** are always independent because they point in completely different directions.

---

## 2. **Vector Dependence**

A set of vectors is **linearly dependent** if **one vector can be expressed as a combination of the others**.

### üëâ Simple meaning

One vector is **redundant**.

### üëâ Geometric meaning

* Dependent vectors lie on the **same line** (collinear) in 2D.
* In 3D, dependent sets lie on the **same plane** or same line.

### üëâ Angle interpretation

* Angle = **0¬∞ or 180¬∞** ‚Üí vectors are parallel.
* Parallel vectors are always **linearly dependent**.

---

## 3. **Determinant Explanation**

When vectors are placed as columns of a square matrix:

[
A = [v_1 \ v_2]
]

### üëâ Independence

[
\det(A) \neq 0
]

* Area (2D) or volume (3D) is **non-zero**.
* Vectors span the full space ‚Üí **independent**.

### üëâ Dependence

[
\det(A) = 0
]

* Area/volume collapses to **zero**.
* Vectors lie in the same direction ‚Üí **dependent**.

### üìå Geometric meaning of determinant

* In **2D**, (|\det(A)|) is the **area of the parallelogram** formed by the vectors.
* In **3D**, (|\det(A)|) is the **volume of the parallelepiped** formed.

If area/volume = 0 ‚Üí vectors don‚Äôt spread out ‚Üí dependent.

---

### 4. **Matrix Inverse Explanation**

A matrix has an inverse **only if** its columns (vectors) are linearly independent.

### üëâ If inverse exists

[
A^{-1} \text{ exists } \iff \det(A) \neq 0 \iff \text{vectors are independent}
]

### üëâ If inverse does NOT exist

[
\det(A) = 0 \iff \text{vectors are dependent}
]

### Explanation

If vectors are dependent, there aren't enough unique directions, so you cannot uniquely solve:

[
A\vec{x} = \vec{b}
]

Thus an inverse cannot exist.

---

### ‚≠ê Final Summary (Easy to Remember)

| Concept            | Independent                                | Dependent                     |
| ------------------ | ------------------------------------------ | ----------------------------- |
| **Angle**          | Angle ‚â† 0¬∞, 180¬∞; orthogonal = independent | Angle = 0¬∞ or 180¬∞ (parallel) |
| **Geometry**       | Span full plane/space                      | Collapse to line or plane     |
| **Determinant**    | det(A) ‚â† 0                                 | det(A) = 0                    |
| **Matrix Inverse** | Exists                                     | Does not exist                |

```python
import numpy as np
import matplotlib.pyplot as plt

# --------- Define vectors ----------
# Linearly independent (not parallel)
v1 = np.array([1, 2])
v2 = np.array([2, 1])

# Orthogonal (90 degrees)
o1 = np.array([2, 0])
o2 = np.array([0, 2])

# Linearly dependent (parallel)
u1 = np.array([1, 2])
u2 = 2 * u1

origin = np.array([[0, 0], [0, 0]])  # common origin for quiver

# --------- Create subplots ----------
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

# 1) Independent (non-parallel) vectors
ax = axes[0]
ax.quiver(*origin, [v1[0], v2[0]], [v1[1], v2[1]], angles='xy', scale_units='xy', scale=1)
ax.set_xlim(-1, 3)
ax.set_ylim(-1, 3)
ax.set_aspect('equal', 'box')
ax.set_title("Linearly Independent")
ax.grid(True)
ax.text(v1[0], v1[1], " v1")
ax.text(v2[0], v2[1], " v2")

# 2) Orthogonal vectors (90¬∞)
ax = axes[1]
ax.quiver(*origin, [o1[0], o2[0]], [o1[1], o2[1]], angles='xy', scale_units='xy', scale=1)
ax.set_xlim(-1, 3)
ax.set_ylim(-1, 3)
ax.set_aspect('equal', 'box')
ax.set_title("Orthogonal (Independent)")
ax.grid(True)
ax.text(o1[0], o1[1], " o1")
ax.text(o2[0], o2[1], " o2")

# 3) Dependent (parallel) vectors
ax = axes[2]
ax.quiver(*origin, [u1[0], u2[0]], [u1[1], u2[1]], angles='xy', scale_units='xy', scale=1)
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 5)
ax.set_aspect('equal', 'box')
ax.set_title("Linearly Dependent")
ax.grid(True)
ax.text(u1[0], u1[1], " u1")
ax.text(u2[0], u2[1], " u2 = 2¬∑u1")

plt.tight_layout()
plt.show()
```

# **Reduced Row Echelon Form (RREF) ‚Äî Short Top-to-Bottom Example**

### **1. Write the augmented matrix**

[
\begin{bmatrix}
1 & -1 & 2
/1 & 2 & 3
\end{bmatrix}
]

### **2. Make zero below the first pivot**

Row2 = Row2 ‚Äì Row1
[
\begin{bmatrix}
1 & -1 & 2 /
0 & 3 & 1
\end{bmatrix}
]

### **3. Make second pivot = 1**

Row2 = Row2 / 3
[
\begin{bmatrix}
1 & -1 & 2 \
/0 & 1 & \tfrac{1}{3}
\end{bmatrix}
]

### **4. Clear above the second pivot**

Row1 = Row1 + Row2
[
\begin{bmatrix}
1 & 0 & \tfrac{7}{3} \
/0 & 1 & \tfrac{1}{3}
\end{bmatrix}
]

### ‚úî This is the RREF.

---

# **Solution**

[
s = \frac{7}{3}, \quad t = \frac{1}{3}
]

---

# **Vector Independence (Using Same Matrix)**

Columns:
[
\begin{bmatrix}
1 & -1 \
/1 & 2
\end{bmatrix}
\rightarrow
\text{RREF} =
\begin{bmatrix}
1 & 0 \
/0 & 1
\end{bmatrix}
]

Both columns have pivots ‚Üí vectors are **linearly independent**.

---

# **Angle Interpretation**

If the angle between vectors is **not** 0¬∞ or 180¬∞, vectors are independent.
Vectors (1,1) and (-1,2) point in different directions ‚Üí **independent**.

---

# **Determinant Link**

[
\det
\begin{bmatrix}
1 & -1 \
/1 & 2
\end{bmatrix}
= 3 \neq 0
]

A non-zero determinant means:

* matrix is invertible
* RREF becomes the identity
* columns are independent
