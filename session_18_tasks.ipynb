````python
# import pandas as pd

VALID_SEAS = ['2022','2021','2020/21','2019','2018','2017','2016','2015',
              '2014','2013','2012','2011','2009/10','2009','2007/08']

def point_table(season):
    if season not in VALID_SEAS:
        raise ValueError(f"Season must be one of {VALID_SEAS}")

    # Filter matches of that season
    s = ipl_matches[ipl_matches['Season'] == season].copy()

    # All teams in this season
    teams = pd.Index(pd.unique(pd.concat([s['Team1'], s['Team2']]))).sort_values()

    # Create empty dataframe
    out = pd.DataFrame(index=teams)
    out.index.name = "TeamName"

    # Matches Played
    mp = pd.concat([s['Team1'], s['Team2']]).value_counts()
    out['MatchesPlayed'] = mp.reindex(teams, fill_value=0)

    # Matches Won
    won = s['WinningTeam'].value_counts()
    out['MatchesWon'] = won.reindex(teams, fill_value=0)

    # No Result (give 1 point to both teams)
    nr = s[s['WonBy'] == 'NoResult'][['Team1','Team2']]
    nr_count = pd.concat([nr['Team1'], nr['Team2']]).value_counts()
    out['NoResult'] = nr_count.reindex(teams, fill_value=0)

    # Points
    out['Points'] = out['MatchesWon'] * 2 + out['NoResult']

    # Sorting: Points descending, TeamName ascending
    out = out.sort_values(['Points','TeamName'], ascending=[False, True])

    return out
````
