### Q1. Given the vectors A = [1, 2, 3] and B = [4, 5, 6], calculate the element-wise product of the two vectors.

````python
A = np.array([1, 2, 3])
B = np.array([4, 5, 6])

print("Element wise product: ", A * B)
````

### Q2. Find the magnitude of the vector C = [2, -3, 1].

````python
C = np.array([2, -3, 1])

# Method 1
magnitude = np.linalg.norm(C)
print(magnitude)

# Method 2
print(np.sqrt(np.sum(C**2)))
````

### Q3. Given the vectors D = [1, 2, 3] and E = [4, 5, 6], calculate the cosine of the angle between them.
cosθ = D⋅E / ∥D∥ * ∥E∥


````python
D = np.array([1, 2, 3])
E = np.array([4, 5, 6])

cosine_angle = np.dot(D, E) / np.linalg.norm(D) * np.linalg.norm(E)
print("Cosine angle between two vectors: ", cosine_angle)
````

### Q4. Given the vectors F = [1, 2, 3] and G = [4, 5, 6], calculate the cross product of the two vectors.

````python
F = np.array([1, 2, 3])
G = np.array([4, 5, 6])

# Method 1
print("Cross product of two vectors is: ", np.cross(F, G))
````

### Q5. Given the vector H = [2, 4, 6], shift the vector +1 unit in each dimension

### Q6. Given the vectors I = [1, 2, 3] and J = [4, 5, 6], calculate the projection of vector I onto vector J.

````python
I = np.array([1, 2 ,3])
J = np.array([4, 5 ,6])

project_i_on_j = (np.dot(I, J) / np.dot(J, J)) * J

print("Projection of I on J: ", project_i_on_j)
````

### Q7. Consider the 2D - position vector a = (1, 3). If b is the point (5, 4).find the vector `ab` and the vector sum `a + ab`. Provide a sketch plot.

Note: Search for quiver, to plot an arrow.

`matplotlib.pyplot.quiver(*args, data=None, **kwargs)`
Plot a 2D field of arrows.

Call signature:

`quiver([X, Y], U, V, [C], /, **kwargs)`
X, Y define the arrow locations, U, V define the arrow directions, and C optionally sets the color. The arguments X, Y, U, V, C are positional-only.

Arrow length

The default settings auto-scales the length of the arrows to a reasonable size. To change this behavior see the scale and scale_units parameters.

Arrow shape

The arrow shape is determined by width, headwidth, headlength and headaxislength. See the notes below.

````python
a = np.array([1, 3])
b = np.array([5, 4])


# Getting vector ab
ab = b - a
print("Vector ab: ", ab)

# Calculating vector sum
a_plus_ab = a + ab
print("Vector a and Vector ab sum is: ", a_plus_ab)

import matplotlib.pyplot as plt

# plotting
origin = np.array([0, 0])
plt.quiver(*origin, a[0], a[1], angles = "xy", scale_units = "xy", scale = 1, color = "r", label = "a")
plt.scatter(b[0], b[1], color = "b", label = "b")
plt.quiver(*a, ab[0], ab[1], angles = "xy", scale_units = "xy", scale = 1, color = "g", label = "ab")
plt.quiver(*origin, a_plus_ab[0], a_plus_ab[1], angles = "xy", scale_units = "xy", scale = 1, color = "y", label = "a+ab")

plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
````


### Q8. Given two 2D vectors `u` and `v` where `u = [-1, 4]` and `v = [3, -2]`, find the dot product `u . v`, the angle between `u` and `v`, and the projection of `u` onto `v`.

````python
u = np.array([-1, 4])
v = np.array([3, -2])

# Dot product
print(f"Dot product of {u} and {v} is {np.dot(u, v)}")

# Angle between
# WRONG FORMULA
# print(f"Angle between {u} and {v} is {(np.dot(u, v))/np.linalg.norm(u) + np.linalg.norm(v)}")
# GIVES COSINE SIMILARITY
# print(f"Angle between {u} and {v} is {np.dot(u, v)/(np.linalg.norm(u) * np.linalg.norm(v))}")
cosine_sim = np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
angle_rad = np.arccos(cosine_sim)
angle_deg = np.degrees(angle_rad)

print(f"Angle between {u} and {v} is {angle_rad} radians")
print(f"Angle between {u} and {v} is {angle_deg} degrees")

# Projection of u on v
print(f"Projection of vector {u} on {v} is {(np.dot(u, v) / np.dot(v, v))*v}")
````

### Q10. Find a nearest point from a given array of points to a point P1.
P1 - (3,9,3)

Array of points :
```
[[1, 8, 5],
[8, 8, 2],
[7, 9, 1],
[5, 6, 3],
[9, 1, 2],
[6, 5, 9],
[5, 8, 6],
[3, 4, 8],
[6, 8, 4],
[7, 2, 7]]
```

````python
p1 = np.array([3, 9, 3])
points = np.array([[1, 8, 5],
[8, 8, 2],
[7, 9, 1],
[5, 6, 3],
[9, 1, 2],
[6, 5, 9],
[5, 8, 6],
[3, 4, 8],
[6, 8, 4],
[7, 2, 7]])

distances = np.linalg.norm(p1 - points)

index_of_min = np.argmin(distances)

point_nearest = points[index_of_min]

print(f"Nearest point to {p1} is {point_nearest}")

# Uncomment below to see plot
# Create figure and axis for 3-D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot all points as blue
ax.scatter(points[:, 0], points[:, 1], points[:, 2], color='blue')

# Plot target point as red
ax.scatter(p1[0], p1[1], p1[2], color='red')

# Plot nearest point as green
ax.scatter(point_nearest[0], point_nearest[1], point_nearest[2], color='yellow')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3-D Scatter Plot')

# Show plot
plt.show()
````
